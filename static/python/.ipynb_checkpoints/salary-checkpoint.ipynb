```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Stroe CSV into DataFrame
salary_file="../../resources/survey_results_public.csv"
#this raw data file is too large to push to the repo so it is removed before pushing to the repo
salary_df = pd.read_csv(salary_file)
salary_df.head()
```

```python
salary_df.columns
```

```python
salary_df['FormalEducation']
```

```python
salary_df_filteredcountry = salary_df[salary_df['Country'] == 'United States']
salary_df_filtered = salary_df_filteredcountry[(salary_df_filteredcountry['Gender'] == 'Female')|(salary_df_filteredcountry['Gender'] == 'Male')]
salary_df_filtered.head()
```

```python
salary_df_filtered_new = salary_df_filtered[['Country', 'Age', 'Gender','Salary','FormalEducation']].copy()
salary_df_filtered_new.head()
```

```python
salary_df_filtered_new.dropna(inplace = True) 
salary_df_filtered_new.head()
```

```python
salary_df_filtered_new.dtypes
```

```python
salary_df_filtered_new["Salary"]= salary_df_filtered_new["Salary"].replace(",", "",regex=True)
```

```python
salary_df_filtered_new[salary_df_filtered_new["Salary"]=="112,000.00"]
```

```python
salary_df_filtered_new['Salary']= salary_df_filtered_new['Salary'].astype(float)
salary_df_filtered_new.dtypes
```

```python
#Save salary data
salary_df_filtered_new.to_csv("../../resources/cleaneddata/salarydata.csv")
salary_df_filtered_new.to_json("../../static/dataset/salary_data.JSON")
```

```python
salary_agegendereducation_median = salary_df_filtered_new.groupby(["Gender","Age","FormalEducation"])['Salary'].median().to_frame().reset_index()

salary_agegendereducation_median.head(10)
```

```python
#Save salary, gender, age, and education data
salary_agegendereducation_median.to_csv("../../resources/cleaneddata/salary_agegenedu_data.csv")
salary_agegendereducation_median.to_json("../../static/dataset/salary_agegenedu_data.JSON")
```

```python
salary_education_median = salary_df_filtered_new.groupby(["FormalEducation"])['Salary'].median().to_frame().reset_index()

salary_education_median.head(10)
```

```python
#Save salar and education data
salary_education_median.to_csv("../../resources/cleaneddata/salary_edu_data.csv")
salary_education_median.to_json("../../static/dataset/salary_edu_data.JSON")
```

```python
salary_age_median = salary_df_filtered_new.groupby(["Age"])['Salary'].median().to_frame().reset_index()

salary_age_median.head(10)
```

```python
#Save age and salary data
salary_age_median.to_csv("../../resources/cleaneddata/salary_age_data.csv")
salary_age_median.to_json("../../static/dataset/salary_age_data.JSON")
```

```python
salary_gender_median = salary_df_filtered_new.groupby(["Gender"])['Salary'].median().to_frame().reset_index()

salary_gender_median.head(10)
```

```python
#Save gender and salary data
salary_gender_median.to_csv("../../resources/cleaneddata/salary_gender_data.csv")
salary_gender_median.to_json("../../static/dataset/salary_gender_data.JSON")
```

```python
salary_education_median["Gender"]=["All"] * len(salary_education_median)
salary_education_median["Age"]=["All"] * len(salary_education_median)

salary_age_median["Gender"]=["All"] * len(salary_age_median)
salary_age_median["FormalEducation"]=["All"] * len(salary_age_median)

salary_gender_median["Age"]=["All"] * len(salary_gender_median)
salary_gender_median["FormalEducation"]=["All"] * len(salary_gender_median)

```

```python
pd_combined = pd.concat([salary_agegendereducation_median, salary_gender_median, salary_age_median, salary_education_median], ignore_index=True, sort=True)
pd_combined.head()
```

```python
pd_combined
```

```python
#Save all calculated salary data
pd_combined.to_csv("../../resources/cleaneddata/salary_median_data.csv")
pd_combined.to_json("../../static/dataset/salary_median_data.JSON")
```
