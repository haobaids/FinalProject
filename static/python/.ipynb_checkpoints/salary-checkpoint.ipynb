```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Stroe CSV into DataFrame
salary_file="../../resources/survey_results_public.csv"
#this raw data file is too large to push to the repo so it is removed before pushing to the repo
salary_df = pd.read_csv(salary_file)
salary_df.head()
```

```python
salary_df.columns
```

```python
salary_df['FormalEducation']
```

```python
salary_df_filtered = salary_df[salary_df['Country'] == 'United States']
salary_df_filtered.head()
```

```python
salary_df_filtered_new = salary_df_filtered[['Country', 'Age', 'Gender','Salary','FormalEducation']].copy()
salary_df_filtered_new.head()
```

```python
salary_df_filtered_new.dropna(inplace = True) 
salary_df_filtered_new.head()
```

```python
salary_df_filtered_new.dtypes
```

```python
salary_df_filtered_new["Salary"]= salary_df_filtered_new["Salary"].replace(",", "",regex=True)
```

```python
salary_df_filtered_new[salary_df_filtered_new["Salary"]=="112,000.00"]
```

```python
salary_df_filtered_new['Salary']= salary_df_filtered_new['Salary'].astype(float)
salary_df_filtered_new.dtypes
```

```python
#Save salary data
salary_df_filtered_new.to_csv("../../resources/cleaneddata/salarydata.csv")
salary_df_filtered_new.to_json("../../static/dataset/salary_data.JSON")
```

```python
salary_df_filtered_new_mean = salary_df_filtered_new.groupby(["Age"])['Salary'].median()

salary_df_filtered_new_mean.head(10)
```

```python
#Save age and salary data
salary_df_filtered_new_mean.to_csv("../../resources/cleaneddata/salary_age_data.csv")
salary_df_filtered_new_mean.to_json("../../static/dataset/salary_age_data.JSON")
```

```python
salary_df_filtered_new_gender = salary_df_filtered_new[(salary_df['Gender'] == 'Male')|(salary_df['Gender'] == 'Female')].groupby(["Gender"])['Salary'].median()

salary_df_filtered_new_gender.head(10)
```

```python
#Save gender and salary data
salary_df_filtered_new_gender.to_csv("../../resources/cleaneddata/salary_gender_data.csv")
salary_df_filtered_new_gender.to_json("../../static/dataset/salary_gender_data.JSON")
```
